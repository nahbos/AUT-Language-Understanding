```python
from datasets import load_dataset
from transformers import AutoConfig, AutoTokenizer, TFAutoModel
```

```python
pquad_dataset = load_dataset('Gholamreza/pquad')
```

```python
pquad_dataset
```

```python
tokenizer = AutoTokenizer.from_pretrained("HooshvareLab/bert-fa-base-uncased")
```

```python
context  = pquad_dataset["train"][0]["context"]
question = pquad_dataset["train"][1]["question"]

tokenizer.decode(tokenizer.encode(question, context))
```
